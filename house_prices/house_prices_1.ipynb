```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
sample_submission = pd.read_csv("../input/house-prices-advanced-regression-techniques/sample_submission.csv")
test = pd.read_csv("../input/house-prices-advanced-regression-techniques/test.csv")
train = pd.read_csv("../input/house-prices-advanced-regression-techniques/train.csv")
```

```python
X_train = train.drop(['SalePrice'], axis=1)
X_test = test
y_train = train['SalePrice'].astype('float32')
```

```python
X_train.dropna(axis=1, inplace = True)
```

```python
X_train.shape
```

```python
features = []
for c in X_train:
    features.append(c)
    print(c)
```

```python
def hist_show(feature_list):
    for c in feature_list:
        li = sorted(train[c].unique())
        f = []
        for v in train[c].unique():
            f.append(train[train[c] == v]['SalePrice'])
        plt.hist(f, histtype="barstacked", label=li)
        plt.legend()
        plt.show()
```

```python
hist_show(['GrLivArea'])
```

```python
hist_show(['OverallQual', 'OverallCond'])
```

```python
hist_show(['ExterQual', 'HeatingQC'])
```

```python
features = ['OverallQual', 'FullBath', 'ExterQual', 'HeatingQC']
```

```python
X_train = X_train[features]
X_test = test[features]
```

```python
X_test.isnull().sum()
```

```python
X_test.head()
```

```python
X_train_num = X_train[['OverallQual', 'FullBath']]
X_test_num = X_test[['OverallQual', 'FullBath']]

X_train_num = pd.concat([X_train_num, pd.get_dummies(X_train['ExterQual'])], axis = 1)
X_train_num = pd.concat([X_train_num, pd.get_dummies(X_train['HeatingQC'])], axis = 1)

X_test_num = pd.concat([X_test_num, pd.get_dummies(X_test['ExterQual'])], axis = 1)
X_test_num = pd.concat([X_test_num, pd.get_dummies(X_test['HeatingQC'])], axis = 1)

X_train_num = X_train_num.astype('float32')
X_test_num = X_test_num.astype('float32')
```

X_train_num.head()

```python
X_test_num.head()
```

```python
X_test_num.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lnr = LinearRegression()
lnr.fit(X_train_num, y_train)

y = lnr.predict(X_test_num)
```

```python
y = pd.DataFrame({'SalePrice': y})
```
